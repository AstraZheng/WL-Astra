```python
sns.get_dataset_names()
crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

```python
# distribution plot
sns.distplot(crash_df['not_distracted'])
```

```python
# joint plot
sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'reg')
```

```python
# kde
sns.kdeplot(crash_df['alcohol'])
```

```python
# pair plot  (numerical values)
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
sns.pairplot(tips_df, hue = 'smoker')
```

```python
sns.rugplot(tips_df['tip'])
```

```python
# styling 
sns.set_style('darkgrid')    # whitegrid; white; dark; 
plt.figure(figsize=(8,4))

sns.set_context('paper', font_scale=1.4)   # "talk", "poster"

sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'reg')
```

```python
# bar plots
sns.barplot(x = 'sex', y = 'total_bill',
            data = tips_df, estimator = np.median) 
              # or mean; variance ..
```

```python
# count plot
sns.countplot(x = 'sex', data = tips_df)
```

```python
# boxplot
sns.boxplot(x = 'day', y = 'total_bill', data=tips_df, hue = 'sex')
plt.legend(loc = 0)
```

```python
# violin plot
sns.violinplot(x = 'day', y = 'total_bill', data = tips_df, hue = 'sex')
```

```python
# strip plot 
sns.stripplot(x = 'day', y = 'total_bill', data = tips_df, jitter = True, hue = 'sex')  
# the scatter plot with categorical data
# dodge = True: separate the hues 
```

#### Palette

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips_df, jitter = True, hue = 'sex', 
            palette='afmhot')
plt.legend(loc = 0)   # upper right: 1; upper left: 2; lower left = 3; lower right = 4  
```

### Matrix Plot

```python
# heat plot 
crash_mx = crash_df.corr()
crash_mx
sns.heatmap(crash_mx, annot = True, cmap="Blues")
# annot = True: show the values
```

```python
flights = sns.load_dataset('flights')
flights
flights = flights.pivot_table(index = 'month', columns= 'year', values = 'passengers')
sns.heatmap(flights, cmap= "Blues")
# df.pivot_table or df.pivot
# fmt = 'd'  purple/red/orange
```

```python
iris = sns.load_dataset('iris')
sns.set_style("white")
iris_g = sns.PairGrid(iris, hue = 'species')   # create empty grids 
iris_g.map_diag(plt.hist)   # or plt.scatter: scatter plot in the diagonal 
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
```

```python
# facet grid
tips_fg = sns.FacetGrid(tips_df, col = 'time', hue = 'smoker')
tips_fg.map(plt.scatter,'total_bill', 'tip')
```

```python
sns.relplot(data=tips_df, x="total_bill", y="tip", hue="size")
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(data=fmri, x="timepoint", y="signal", hue = "event", kind="line")   
# can use errorbar = None  to remove the bands
```

```python
sns.relplot(
    data=fmri.query("region == 'frontal'"), kind="line",
    x="timepoint", y="signal", hue="event", style="event",
    col="subject", col_wrap=5,
    height=3, aspect=.75, linewidth=2.5,
)
```

Summary: sns.relplot(dataset, x = , y = , kind = , hue = , size = , palette = , col = , row = , col_wrap = , ... )

### sns.displot
For the distribution 

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.displot(penguins, x="flipper_length_mm", bins = 20, shrink = 0.9)    
# bins: the # of bins; different from binwidth
# shrink: the gaps between bars 
# y axis: count
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", element="step")
# element = "step"
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", multiple = "stack")
```

```python
sns.displot(penguins, x="flipper_length_mm", 
            hue="species", hue_order=['Adelie', 'Gentoo','Chinstrap'], 
            multiple = "stack")
# hue_order \
```

normalized histogram: stat = "density"

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", stat="density", common_norm = False, alpha = 0.3) 
# y axis: the density 
# alpha: the opacity 
```

```python
sns.displot(penguins, x="flipper_length_mm", kind="kde", bw_adjust=.25)
# bw_adjust: bandwidth
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", kind="kde", fill=True)
```
